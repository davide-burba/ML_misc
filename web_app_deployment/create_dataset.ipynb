```python
from google_images_download import google_images_download
```

We will use package from https://google-images-download.readthedocs.io/en/latest/index.html

**Note:** to download more than 100 images per search, you need chrome and chromedriver (https://sites.google.com/a/chromium.org/chromedriver/downloads)

```python
# set arguments
keywords = "windmill breakdance,thomas breakdance,air flare breakdance"
limit = 200

path_to_chromedriver = '/Applications/chromedriver'
```

```python
#class instantiation
response = google_images_download.googleimagesdownload()   

#creating list of arguments
arguments = {
    "keywords":keywords,
    "limit":limit,
    "print_urls":False,
    'chromedriver':path_to_chromedriver,
}  
```

```python
paths = response.download(arguments)   #passing the arguments to the function
```
